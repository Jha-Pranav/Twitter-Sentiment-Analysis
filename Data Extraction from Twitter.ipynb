```python
# Importing Twitter module
from tweepy import OAuthHandler, API
import pandas as pd
```

```python
# Twitter Credentials
consumer_key = 'xxxxxxxxxxxxxxxxxxx'
consumer_secret = 'xxxxxxxxxxxxxxxxxxxxxxx'
access_token = 'xxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxx'
access_token_secret = 'xxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxx'
```

```python
# Authentication
try:
    auth = OAuthHandler(consumer_key, consumer_secret
except:
    print("Error: Authentication Failed") 
```

```python
# Setting Access token
try:
    auth.set_access_token(access_token,access_token_secret)
except:
    print("Error: Wrong Access token") 
```

```python
# Creating API to fetch tweets
api = API(auth)
```

```python
# Fetching Tweets 
collection  = []
# Query Keywords for gender dataset
# query = ['gender','male','female','femininity','sexuality','womanhood','womanliness','manliness','manhood','masculine','androgynous']

# Query keywords for  election 2020
# query = ['2020election','DelhiElections2020','delhiassemblyelections2020','kejriwal','Arvindkejriwal','Modi','rahulgandhi','Aap','congress','congressmuktbharat','evm','exitpool','delhielection','opiniopoll','amitshah','bjp']

query = ['Shikara']
for q in query:
    for lang in ['hi','en']:
        fetched_tweets = api.search(q=q,lang=lang,count=100,result_type='recent')
        collection.append(fetched_tweets)
```

```python
tweets = []
tweet_id = []
lang = []
location = []
for coll in collection:
    for tweet in coll:
        if tweet._json['text'] not in tweets:   # Avoid Duplication , copied posts
            tweets.append(tweet._json['text'])
            tweet_id.append(tweet._json['id'])
            lang.append(tweet._json['lang'])
            location.append(tweet._json['user']['location'])

data = {'Id': tweet_id,'Tweets': tweets,'Lang': lang, 'Location': location}
```

```python
df = pd.DataFrame(data,columns=['Id','Tweets','Lang','Location'])
```
